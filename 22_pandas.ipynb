```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
#나이에 따라 직원을 분류하는 함수 정의
#young , semi_mid, mid

def age_value(age):
    if age < 30:
        return 'young'
    elif age < 40:
        return'semi_mid'
    else:
        return 'mid'
```

```python
#데이터프레임에 나이에 따라 직원 분류 함수 적용해 catgegory 컬럼 추가
df['Category'] = df['Age'].apply(age_value)
```
