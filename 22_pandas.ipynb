## 2.4 열추가

```python
sci2.info()
```

```python
#str - > datetime
born_datetime = pd.to_datetime(sci2['Born'], format='%Y-%m-%d' )
born_datetime
```

```python
sci2['born_dt'] = born_datetime
sci2.info()
```

```python
died_datetime = pd.to_datetime(sci2['Died'], format='%Y-%m-%d')

sci2['died_dt'] = died_datetime
```

```python
sci2
```

```python
sci2['age_days'] = sci2['died_dt'] - sci2['born_dt']
sci2['age_days'].dtype
```

```python
#series.apply(함수)
import numpy as np
sci2['age_days'] = (sci2['age_days'].dt.days / 365).apply(np.floor)
```

```python
sci2
```

```python
# pd.concat([first_half, second]) # 합치기
```

```python
#df.assign( col_name = 수식) 적용해

sci2 = sci2.assign(
    age_days_assign = sci2['died_dt'] - sci2['born_dt'] ,
    age_assign = ((sci2['died_dt'] - sci2['born_dt']).dt.days / 365).apply(np.floor)
)

    #age_assign = ((sci2['age_days'].dt.days / 365).apply(np.floor)
```
